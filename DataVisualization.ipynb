# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Number of images per class

```python
df = pd.read_csv('annot_tot.csv')

df.head()
```

```python
np.unique(df.ok_check)
```

```python
fig, ax = plt.subplots(figsize = (18, 10))

sns.countplot(data = df, x = 'class', 
              ax = ax, order = df['class'].value_counts().index)
plt.xticks(rotation = 45)
plt.title('Number of images per logo, pre cleaning', fontsize = 20)
plt.show()
```

## Number of images per class post data cleaning

```python
np.unique(df['class'])
```

```python

df1 = df[(df.ok_check == 1) | (df.ok_check.isna())]

logos_kept = ['Adidas', 'AppleInc',  'CocaCola', 
       'HardRockCafe', 'MercedesBenz', 'NFL', 'Nike',
       'Puma', 'Starbucks',
       'TheNorthFace', 'UnderArmour']

df1 = df1[df['class'].isin(logos_kept)]
fig, ax = plt.subplots(figsize = (18, 10))

sns.countplot(data = df1, x = 'class', 
              ax = ax, order = df1['class'].value_counts().index)
plt.xticks(rotation = 45)
plt.title('Number of images per logo, post cleaning', fontsize = 20)
plt.show()
```
